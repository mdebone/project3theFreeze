```python
from bs4 import BeautifulSoup as bs
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import re
import time 
from datetime import datetime
```

```python
#set up chrome browser
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#define empty dataframe to append data to
reuters_news = pd.DataFrame(columns = ['timestamp', 'newstext', 'link'])
keyword_list=['Texas', 'storm', 'winter', 'ice', 'freeze', 'blackout', 'power', 'outage', 'electicity', 'cold', 'icy', 'electric']

#define function to convert reuters text string to string suitable for strptime
def correct_date(a):
    r = a.split(" ")
    del r[6]
    d = r[4].split(":")
    d[0] = d[0].zfill(2)
    d[1] = d[1].zfill(2)
    r[4] = ":".join(d)
    return " ".join(r)





#define a subroutine to extract data from house structure and append it to a dataframe
def extract_from_news_item(item):
    global reuters_news
    item_text=item.find_all('span')[1].text
    item_time=item.find('time').text.replace('Â·', '')
    item_href='https://www.reuters.com' + item.a['href']
    item_time_corrected=correct_date(item_time)
    datetime_obj=datetime.strptime(item_time_corrected, "%B %d, %Y  %I:%M %p")
    time_stamp=datetime_obj.strftime('%Y-%m-%d %H:%M:%S.%f')[:-3]
    if any(x in item_text for x in keyword_list):
        reuters_news = reuters_news.append({'timestamp':time_stamp, 'newstext':item_text, 'link':item_href}, ignore_index = True) 

for pagenum in range(420, 600, 10):
    print(pagenum)
    #DEFINE THE URL PAGE
    url=f'https://www.reuters.com/site-search/?query=Texas&offset={pagenum}&sort=oldest&date=past_year'
    browser.visit(url)
    time.sleep(5)
    html = browser.html
    soup = bs(html, 'html.parser')
    #Scrape the news items for the current search page
    news_items=soup.find_all('li', class_="SearchResults__item___3jzYEE")
    for item in news_items:
        extract_from_news_item(item)
        
print('Done!')
```
